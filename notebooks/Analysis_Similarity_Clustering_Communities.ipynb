```python
idx_df = gov_df.set_index('Name')
idx_df.head()
```

```python
gov_num_df = idx_df.select_dtypes(include='number')
```

```python
gov_num_df.head()
```

```python
## StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_features = StandardScaler().fit_transform(gov_num_df.values)
```

```python
scaled_features
```

```python
scaled_df = pd.DataFrame(scaled_features, index=gov_num_df.index, columns=gov_num_df.columns)
```

```python
scaled_df.head()
```

## KMeans Clustering

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(scaled_df)
```

```python
clusters = kmeans.predict(scaled_df)
```

```python
scaled_df['k_cluster'] = clusters
```

```python
scaled_df['k_cluster']
```

```python
scaled_df.head()
```

### KMeans + PCA

```python
pca = PCA(n_components=2)
```

```python
pca_df = pd.DataFrame(
    pca.fit_transform(scaled_df.drop(['k_cluster'], axis=1)),
    index=idx_df.index
)
```

```python
pca_df.columns = ['pca_1', 'pca_2']
```

```python
pca_df.head()
```

```python
plot_df = pd.concat(
    [scaled_df, pca_df], 
    axis=1, 
    join='outer', 
)
```

```python
plot_df.head()
```

```python
cluster_0 = plot_df[plot_df['k_cluster'] == 0]
cluster_1 = plot_df[plot_df['k_cluster'] == 1]
cluster_2 = plot_df[plot_df['k_cluster'] == 2]
```
