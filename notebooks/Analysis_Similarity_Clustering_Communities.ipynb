```python
idx_df = gov_df.set_index('Name')
idx_df.head()
```

```python
gov_num_df = idx_df.select_dtypes(include='number')
```

```python
gov_num_df.head()
```

### Normalize

```python
scaler = MinMaxScaler()
```

```python
scaled_features = scaler.fit_transform(gov_num_df.values)
```

```python
scaled_features
```

```python
scaled_df = pd.DataFrame(scaled_features, index=gov_num_df.index, columns=gov_num_df.columns)
```

```python
scaled_df.head()
```

# KMeans Clustering

```python
kmeans = KMeans(n_clusters=4, random_state=seed)
```

```python
kmeans.fit(scaled_df)
```

```python
k_clusters = kmeans.predict(scaled_df)
```

```python
k_clusters_df = pd.DataFrame(
    k_clusters, 
    index=idx_df.index, 
    columns=['k_cluster']
)
```

```python
k_clusters_df
```

### KMeans + PCA

```python
pca = PCA(n_components=2)
```

```python
pca_df = pd.DataFrame(
    pca.fit_transform(scaled_df),
    index=idx_df.index
)
```

```python
pca_df.columns = ['pca_1', 'pca_2']
```

```python
pca_df.head()
```

```python
plot_df = pd.concat(
    [scaled_df, k_clusters_df, pca_df], 
    axis=1, 
    join='outer', 
)
```

```python
plot_df.head()
```

```python
k_cluster_0 = plot_df[plot_df['k_cluster'] == 0]
k_cluster_1 = plot_df[plot_df['k_cluster'] == 1]
k_cluster_2 = plot_df[plot_df['k_cluster'] == 2]
k_cluster_3 = plot_df[plot_df['k_cluster'] == 3]
```

```python
k_cluster_0.head()
```

### Visualize PCA 

```python
init_notebook_mode(connected=True)
```

```python
[print(rowname) for rowname in plot_df.index]
```

```python
trace_0 = go.Scatter(
    x=k_cluster_0['pca_1'],
    y=k_cluster_0['pca_2'],
    mode='markers',
    text=k_cluster_0.index,
    marker=dict(color='purple'),
    name='Cluster 0'
)

trace_1 = go.Scatter(
    x=k_cluster_1['pca_1'],
    y=k_cluster_1['pca_2'],
    mode='markers',
    text=k_cluster_1.index,
    marker=dict(color='orange'),
    name='Cluster 1'
)

trace_2 = go.Scatter(
    x=k_cluster_2['pca_1'],
    y=k_cluster_2['pca_2'],
    mode='markers',
    text=k_cluster_2.index,
    marker=dict(color='green'),
    name='Cluster 2'
)

trace_3 = go.Scatter(
    x=k_cluster_3['pca_1'],
    y=k_cluster_3['pca_2'],
    mode='markers',
    text=k_cluster_3.index,
    marker=dict(color='pink'),
    name='Cluster 3'
)

title = 'K-Means + PCA Clustering for Community Similarity'

data = [trace_0, trace_1, trace_2, trace_3]
layout = dict(title=title,
              xaxis=dict(title= 'pc1',ticklen=5,zeroline=False),
              yaxis=dict(title= 'pc2',ticklen=5,zeroline=False)
             )

fig = dict(data=data, layout=layout)

iplot(fig)
```

# DBSCAN

```python
dbscan = DBSCAN(eps=3, min_samples=2).fit(scaled_df)
```

```python
dbscan
```

```python
labels = dbscan.labels_
```

```python
# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```

```python
dbscan_df = pd.DataFrame(
    labels,
    index=idx_df.index,
    columns=['dbscan']
)
```

```python
dbscan_df
```

```python
dbscan_df.value_counts()
```

## DBSCAN + PCA
